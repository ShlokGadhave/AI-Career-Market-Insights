```python
llm_counts=df[df['is_llm_role']==1]
skills_with_llm=llm_counts.groupby('required_skills').size()
skills_with_llm.head(10)
```

```python
df.columns
llm_jobs = df[df['is_llm_role'] == 1]
llm_jobs

from collections import Counter
skill_count = Counter()
for skills in llm_jobs['required_skills'].dropna():
    skills_list = [skill.strip() for skill in skills.split(',')]
    skill_count.update(skills_list)

skill_count.most_common(10)

# Conversion Into DataFrame
skills_df = pd.DataFrame(
    skill_count.items(),
    columns=['Skill', 'Count']
)
skills_df = skills_df.sort_values(
    by='Count',
    ascending=False
)
skills_df.head(10)
```

```python
entry_jobs = df[df['experience_level'] == 'Entry (0-2 yrs)']
entry_jobs

from collections import Counter
skill_count = Counter()
for skills in entry_jobs['required_skills'].dropna():
    skills_list = [skill.strip() for skill in skills.split(',')]
    skill_count.update(skills_list)

skill_count.most_common(10)

skills_df = pd.DataFrame(
    skill_count.items(),
    columns=['Skill', 'Count']
)
skills_df = skills_df.sort_values(
    by='Count',
    ascending=False
)
skills_df.head(10)

```

```python

top_skills = skills_df.head(10)

plt.figure(figsize=(10,6))

sns.barplot(
    data=top_skills,
    x='Count',
    y='Skill',
    palette='RdBu'
)

plt.title('Top Skills Required for Entry-Level Jobs')
plt.xlabel('Frequency')
plt.ylabel('Skill')

plt.show()
```

```python
senior_jobs = df[df['experience_level'] == 'Senior (6-9 yrs)']
senior_jobs

from collections import Counter
skill_count = Counter()
for skills in senior_jobs['required_skills'].dropna():
    skills_list = [skill.strip() for skill in skills.split(',')]
    skill_count.update(skills_list)
skill_count.most_common(10)

skills_df = pd.DataFrame(
    skill_count.items(),
    columns=['Skill', 'Count']
)
skills_df = skills_df.sort_values(
    by='Count',
    ascending=False
)
skills_df.head(10)

```

```python


top_skills = skills_df.head(10)

plt.figure(figsize=(10,6))

sns.barplot(
    data=top_skills,
    x='Count',
    y='Skill',
    palette='RdBu'
)

plt.title('Top Skills Required for Entry-Level Jobs')
plt.xlabel('Frequency')
plt.ylabel('Skill')

plt.show()
```
