```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df['avg_salary']=df['salary_min_usd']+df['salary_max_usd']/2
df[['job_title','avg_salary']].head(10)
```

Salary Analysis 


```python
df.groupby('job_title')['avg_salary'].max().sort_values(ascending=False)
#AI Jobs with Highest Salary is AI Solutions Architect
```

```python
df.groupby('job_title')['avg_salary'].min().sort_values(ascending=True)
#AI Business Analyst is the Lowest Salary AI Job 
```

```python
df.groupby('country')['avg_salary'].mean().sort_values(ascending=False)
#France is country having highest salaries
```

```python
df.groupby('city')['avg_salary'].mean().sort_values(ascending=False)
#Paris is the city having Highest Salary
```

```python
exp_salary=df.groupby('years_of_experience')['annual_salary_usd'].mean().sort_values(ascending=False)
exp_salary.head(1)
exp_salary.tail(1)
# Yes as in 13 year has the maximum the salary whereas experience of 1 year has the lowest salary
```

```python
size_vs_package=df.groupby('company_size')['avg_salary'].mean()
size_vs_package.head(1)
size_vs_package.tail(1)
#Big tech companies have the avg salaries and startup have the avg salaries

```
