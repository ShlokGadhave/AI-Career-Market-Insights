```python
df['avg_salary']=df['salary_min_usd']+df['salary_max_usd']/2
df[['job_title','avg_salary']].head(10)
```

Salary Analysis 


```python
df.groupby('job_title')['avg_salary'].max().sort_values(ascending=False)
#AI Jobs with Highest Salary is AI Solutions Architect
```

```python
df.groupby('job_title')['avg_salary'].min().sort_values(ascending=True)
#AI Business Analyst is the Lowest Salary AI Job 
```

```python
df.groupby('country')['avg_salary'].mean().sort_values(ascending=False)
#France is country having highest salaries
```

```python
sns.boxplot(x='years_of_experience',y='annual_salary_usd',data=df)
plt.title('Yrs of Experience compare to salary')
```

Work Mode Insights

```python

pd.pivot_table(df,
               values='remote_work',
               index='country',
               aggfunc='count')
```

```python
sns.boxplot(x='remote_work',y='country',data=df)
plt.title('Count Work as per country')
```

```python
count_remote = df[df['remote_work'] == 'Fully Remote']
remote_country=count_remote.groupby('country').size().sort_values(ascending=False).reset_index(name='remote_work')
remote_country
#USA has the highest count of remote jobs 
```

```python
sns.barplot(
    data=remote_country,
    x='remote_work',
    y='country',
    palette='viridis'
)
plt.title('Top Countries with Fully Remote Jobs')
plt.xlabel('Number of Remote Jobs')
plt.ylabel('country')

plt.xticks(rotation=45)

plt.show()
```

```python
count_remote = df[df['remote_work'] == 'Fully Remote']
remote_jobs=count_remote.groupby('job_title').size().sort_values(ascending=False).reset_index(name='remote_work')
remote_jobs.head(10)
#AI Product manager has the highest remote jobs
```

```python
sns.barplot(data=remote_jobs,x='remote_work',y='job_title',palette='RdBu')
plt.title('AI Jobs disturbution with Number of Remote Work')
plt.xlabel('AI Jobs')
plt.ylabel('Remote Work')
```

```python
remote_salariers=df.groupby('remote_work')['annual_salary_usd'].mean().sort_values(ascending=False).reset_index()
remote_salariers

# As per the analysis remote work as per analysis has the better annual salary/package
```

```python
count_hybrid=df[df['remote_work']=='Hybrid']
hybrid_work=count_hybrid.groupby('industry').size().sort_values(ascending=False).reset_index(name='remote_work')
hybrid_work
# Consulting industry provides highest hybrid work 

```

```python
df.columns
df['remote_work']
count_jobs=df.groupby('remote_work').size().sort_values(ascending=False).reset_index(name='job_count')
count_jobs
```

```python
sns.barplot(
    data=count_jobs,
    x='remote_work',
    y='job_count',
    palette='viridis'
)
plt.title('Distubrution btw onsite vs remote jobs ')
plt.xlabel('Remote Work')
plt.ylabel('Job Count')
```

```python
remote_flexibility=df.groupby('company_size')['is_remote_friendly'].size().sort_values(ascending=False).reset_index(name='count of remote flexbiliy')
remote_flexibility
# Mid-size comapnies offers better flexibility of remote work 
```

```python
remotework_vs_exp=df.groupby('remote_work')['experience_level'].size().sort_values(ascending=False).reset_index(name='job_count')
remotework_vs_exp
# As compare to experience level hybrid jobs gets more opporunties 


```

```python
sns.barplot(data=remotework_vs_exp,x='remote_work',y='job_count',palette='plasma')
plt.title('Work Mode Insights based on experience')
plt.xlabel('Count of Experience')
plt.ylabel('Work Modes')
```
