```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df['avg_salary']=df['salary_min_usd']+df['salary_max_usd']/2
df[['job_title','avg_salary']].head(10)
```

```python
sns.barplot(x='avg_salary',y='job_title',data=df,color='red')
plt.title('avg salary of each AI Jobs ')
```

Salary Analysis 


```python
df.groupby('job_title')['avg_salary'].max().sort_values(ascending=False)
#AI Jobs with Highest Salary is AI Solutions Architect
```

```python
df.groupby('job_title')['avg_salary'].min().sort_values(ascending=True)
#AI Business Analyst is the Lowest Salary AI Job 
```

```python
df.groupby('country')['avg_salary'].mean().sort_values(ascending=False)
#France is country having highest salaries
```

```python
sns.histplot(x='avg_salary',y='country',data=df)
plt.title('avg salary of each country')
```

```python
df.groupby('city')['avg_salary'].mean().sort_values(ascending=False)
#Paris is the city having Highest Salary
```

```python
exp_salary=df.groupby('years_of_experience')['annual_salary_usd'].mean().sort_values(ascending=False)
exp_salary.head(1)
exp_salary.tail(1)
# Yes as in 13 year has the maximum the salary whereas experience of 1 year has the lowest salary
```

```python
sns.boxplot(x='years_of_experience',y='annual_salary_usd',data=df)
plt.title('Yrs of Experience compare to salary')
```

```python
size_vs_package=df.groupby('company_size')['avg_salary'].mean()
size_vs_package.head(1)
size_vs_package.tail(1)
#Big tech companies have the avg salaries and startup have the avg salaries

```

```python
sns.boxplot(x='avg_salary',y='company_size',data=df)
plt.title('size_vs_package')
```

```python
#Salary Disturbution of AI Jobs 
df.groupby('job_title')['avg_salary'].mean()
```

```python
df.groupby('job_category')['avg_salary'].mean().sort_values(ascending=False)
# Architecture has the highest employement type as per salary
```

```python
#Salary Differences btw various job modes 
# df.groupby('remote_work')['annual_salary_usd'].mean()
pd.pivot_table(df,
               values='annual_salary_usd',
               index='remote_work',
               aggfunc='mean')

```

```python
#Top AI Jobs
top_jobs = (
    df.groupby('job_title')['annual_salary_usd']
    .mean()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)
top_jobs
```

```python
sns.barplot(data=top_jobs,x='annual_salary_usd',y='job_title')
```

```python
pd.pivot_table(df,values='annual_salary_usd',index='industry',aggfunc='mean')
#Automotive industires pay the most to the employees 

```

```python
sns.barplot(x='annual_salary_usd',y='industry',data=df)
plt.title('Annual Salary in various Job Roles')
```

```python

education_vs_salary=df.groupby('education_required')['annual_salary_usd'].mean().sort_values(ascending=False).reset_index()
education_vs_salary

# education affects the salary as Phd ones has the highest salaries in the list where as associates has the lowest salaries 
```

```python
sns.barplot(data=education_vs_salary,x='education_required',y='annual_salary_usd')
plt.title('Education Required vs Salary')
```

```python
variation = df.groupby('job_title')['annual_salary_usd'].agg(['min','max'])
variation['salary_gap'] = variation['max'] - variation['min']
variation.sort_values('salary_gap', ascending=False)

```
